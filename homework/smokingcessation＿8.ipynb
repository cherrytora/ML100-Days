```python
# 資料重新排序
data0 = cessation_8[['No','H_0','BA_0','BD_0']]
data0['Times'] = 0
data0['TimeDiff0'] = 0
data0 = data0.rename(columns = {'H_0':'hos','BA_0':'BA','BD_0':'BD'})

data1 = cessation_8[['No', 'H_1' ,'BA_1','BD_1','Time_1-Time0']]
data1['Times'] = 1
data1 = data1.rename(columns = {'H_1':'hos','BA_1':'BA','BD_1':'BD'})
data1 = data1.rename(columns = {'Time_1-Time0': 'TimeDiff0'})

data2 = cessation_8[['No','H_2','BA_2','BD_2','Time_2-Time0']]
data2['Times'] = 2
data2 = data2.rename(columns = {'H_2':'hos','BA_2':'BA','BD_2':'BD'})
data2 = data2.rename(columns = {'Time_2-Time0': 'TimeDiff0'})

data3 = cessation_8[['No','H_3','BA_3','BD_3','Time_3-Time0']]
data3['Times'] = 3
data3 = data3.rename(columns = {'H_3':'hos','BA_3':'BA','BD_3':'BD'})
data3 = data3.rename(columns = {'Time_3-Time0': 'TimeDiff0'})
data = pd.concat([data0,data1,data2,data3],ignore_index=True)


cols=['No','Times','TimeDiff0','hos','BA','BD']
data=data.ix[:,cols]

# 後側和前測分數差異＿敵意量表
cessation_8_clean = cessation_8[['No','Time_1-Time0','Time_2-Time0','Time_3-Time0']] 
cessation_8_clean['H1-H0']=cessation_8['H_1']-cessation_8['H_0']
cessation_8_clean['H2-H0']=cessation_8['H_2']-cessation_8['H_0']
cessation_8_clean['H3-H0']=cessation_8['H_3']-cessation_8['H_0']
```
